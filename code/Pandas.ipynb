#### 数据读取

```python
import pandas as pd
df = pd.read_csv('data/usa_china_jpn_gdp_1960_2022.csv', sep = ',')
df
```

#### plot绘图

```python
china_gdp = df[df['Country'] == 'China'].set_index('Year')
us_gdp    = df[df['Country'] == 'USA'].set_index('Year')
jp_gdp    = df[df['Country'] == 'Japan'].set_index('Year')

china_gdp.GDP_USD_Bil.plot()
us_gdp.GDP_USD_Bil.plot()
jp_gdp.GDP_USD_Bil.plot()
```

#### 设置图例

```python
china_gdp = df[df['Country'] == 'China'].set_index('Year')
us_gdp    = df[df['Country'] == 'USA'].set_index('Year')
jp_gdp    = df[df['Country'] == 'Japan'].set_index('Year')

china_gdp.GDP_USD_Bil.plot(legend = True)
us_gdp.GDP_USD_Bil.plot(legend = True)
jp_gdp.GDP_USD_Bil.plot(legend = True)
```

#### 修改列名间接修改图例

```python
china_gdp = df[df['Country'] == 'China'].set_index('Year')
us_gdp    = df[df['Country'] == 'USA'].set_index('Year')
jp_gdp    = df[df['Country'] == 'Japan'].set_index('Year')

china_gdp.rename(columns={"GDP_USD_Bil":"China"} ,inplace = True)
jp_gdp.rename(columns={"GDP_USD_Bil":"Japan"} ,inplace = True)
us_gdp.rename(columns={"GDP_USD_Bil":"USA"} ,inplace = True)

china_gdp.China.plot(legend = True)
us_gdp.USA.plot(legend = True)
jp_gdp.Japan.plot(legend = True)
```

```python
china_gdp = df[df['Country'] == 'China'].set_index('Year')['GDP_USD_Bil']
us_gdp    = df[df['Country'] == 'USA'].set_index('Year')['GDP_USD_Bil']
jp_gdp    = df[df['Country'] == 'Japan'].set_index('Year')['GDP_USD_Bil']

ax = china_gdp.plot(label='China', figsize=(9, 5), title='GDP over Time')
us_gdp.plot(ax=ax, label='USA')
jp_gdp.plot(ax=ax, label='Japan')

ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('GDP (Billion USD)')
```
