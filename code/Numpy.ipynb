### Numpy属性

```python
import numpy as np

a = np.arange(15).reshape(3, 5)
# arange,生成一维矩阵
print('矩阵形状:', a.shape)
print('矩阵维数：', a.ndim)
print('矩阵元素类型:', a.dtype)
print('矩阵元素大小:', a.itemsize)
print('矩阵大小:', a.size)
print('矩阵类型:', type(a))
```

### ndarray

```python
a = np.array([2, 3, 4])
print("数组a元素类型:",a)
print("数组a类型:", a.dtype)

b = np.array([1.2, 3.5, 5.1])
print("数组b元素类型:", b.dtype)
```

```python
np_arange =np.arange(10, 20, 5, dtype=int)
print("arange创建np_arange:", np_arange)
print("arange创建np_arange的元素类型:", np_arange.dtype)
print("arange创建np_arange的类型:", type(np_arange))
```

### 随机数矩阵

```python
arr = np.random.rand(3, 4) 
print(arr)
print(type(arr)) 

arr = np.random.randint(-1, 5, size=(3, 4))
print(arr)
print(type(arr))

arr =np.random.uniform(-1, 5, size=(3, 4))
print(arr)
print(type(arr))
```

### 类型转换

```python
# 初始化3行4列数组，数据类型为f1oat64
zeros_float_arr = np.zeros((3,4),dtype=np.float64)
print(zeros_float_arr)
print(zeros_float_arr.dtype) #float64
# astype转换数据类型，将已有的数组的数据类型转换为int32
zeros_int_arr = zeros_float_arr.astype(np.int32)
print(zeros_int_arr)
print(zeros_int_arr.dtype) #int32
```

### numpy内置函数

#### 基本函数

```python
arr = np.random.randn(2,3)
print(arr)
```

```python
print(np.ceil(arr))
print(np.floor(arr))
print(np.rint(arr))
print(np.isnan(arr))
print(np.multiply(arr, arr))
print(np.divide(arr, arr))
print(np.where(arr>0, 1, -1))
```

#### 统计函数

```python
arr = np.arange(12).reshape(3, 4)
print(arr)
```

```python
print(np.mean(arr))
print(np.mean(arr, axis=0))
print(np.mean(arr, axis=1))
print(np.sum(arr))
print(np.sum(arr, axis=0))
print(np.sum(arr, axis=1))
print(np.max(arr))
print(np.max(arr, axis=0))
print(np.max(arr, axis=1))
print(np.min(arr))
print(np.min(arr, axis=0))
print(np.min(arr, axis=1))
print(np.var(arr))
print(np.var(arr, axis=0))
print(np.var(arr, axis=1))
```

```python
print(np.std(arr))
print(np.std(arr, axis=0))
print(np.std(arr, axis=1))
print(np.argmin(arr))
print(np.argmin(arr, axis=0))
print(np.argmin(arr, axis=1))
print(np.argmax(arr))
print(np.argmax(arr, axis=0))
print(np.argmax(arr, axis=1))
print(np.cumsum(arr))
print(np.cumsum(arr, axis=0))
print(np.cumsum(arr, axis=1))
print(np.cumprod(arr))
print(np.cumprod(arr, axis=0))
print(np.cumprod(arr, axis=1))
```
